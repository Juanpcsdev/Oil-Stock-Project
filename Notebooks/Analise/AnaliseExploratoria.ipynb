```python
#Criação de uma coluna ano/mes
df1['ano_mes'] = df1['Date'].dt.to_period('M')
```

```python
df1.head(2)
```

```python
#Selecionando apenas os valores de fechamentos mensais

colunas_media = ['Close_Real_EYAC2','Close_Dolar_EYAC2','Volume_EYAC2','Price_Dolar_Petro','Volume_Petro','Índice_Ibovespa','ano_mes']

list_dates = []

#Adicionando valores do fechamento (último dia) de cada mês
for x in range (2010,2025):
    for y in range (1,13):
        list_dates.append(list(df1[(df1['Date'].dt.year==x) & (df1['Date'].dt.month == y)].iloc[-1]))

#Transformando em dataframe com colunas selecionadas
df_mean = pd.DataFrame(list_dates, columns=df1.columns)
df_mean = df_mean[colunas_media]
```

```python
#União da tabela de fechamento com os valores de Inflação mensais
df_mean = df_mean.merge(df1[['ano_mes','Value_IPCA']].drop_duplicates(),right_on='ano_mes', left_on='ano_mes',how='left')
```

```python
df_mean.tail(2)
```

```python
# df_mean['Change%_Petro'] = df_mean['Price_Dolar_Petro'].pct_change() * 100
# df_mean['Change%_Ibovespa'] = df_mean['Índice_Ibovespa'].pct_change() * 100
```

```python
#Valor da inflação como tipo float
df_mean['Inflac_IPCA'] = df_mean['Value_IPCA'].replace({'%':''},regex=True).astype(float)
```

```python
#Adicionando coluna de inflação acumulada em 12 meses

lista_ipca = []

for i in df_mean.index:
    taxas_inflacao = [x for x in df_mean['Inflac_IPCA'].iloc[max(0, i-12):i] / 100]
    #Calculando a inflação acumulada
    inflacao_acumulada = 1
    for taxa in taxas_inflacao:
        inflacao_acumulada *= (1 + taxa)

    inflacao_acumulada -= 1
    inflacao_acumulada = inflacao_acumulada * 100  #Convertendo para porcentagem
    lista_ipca.append(inflacao_acumulada)

df_mean['IPCA_Acumulado'] = lista_ipca
```

```python
#Transformando coluna ano/mes em timestamp
df_mean['Date'] = df_mean['ano_mes'].dt.to_timestamp()
```
