1. Influência Direta: Ações Ligadas ao Setor de Petróleo
• O Ibovespa é composto por várias empresas, e Petrobras (PETR3, PETR4) tem um peso enorme no índice.
• Como a Petrobras é uma petroleira, o preço do barril de petróleo influencia diretamente seu lucro.
• Alta no petróleo → tende a aumentar lucro da Petrobras → ações sobem → Ibovespa sobe.
• Queda no petróleo → pode reduzir lucro → ações caem → impacto negativo no Ibovespa.
⸻
2. Relação com o Câmbio
• O petróleo é cotado em dólar, e o Brasil é um exportador de petróleo.
• Quando o petróleo sobe, pode haver:
• Entrada de dólares no Brasil (via exportações)
• Valorização do real (ou menor pressão de desvalorização)
• Isso pode impactar positivamente o Ibovespa, especialmente empresas que se beneficiam de dólar mais baixo (como varejo e consumo interno).
⸻
3. Impacto na Inflação e Juros
• O petróleo afeta o custo dos combustíveis (gasolina, diesel), que influenciam o IPCA (inflação).
• Alta no petróleo → aumento da inflação → pode levar o Banco Central a subir juros → impacto negativo no Ibovespa.
• Então, o petróleo também afeta o ciclo de política monetária indiretamente.
⸻
4. Relação Global e Sentimento de Mercado
• Alta no petróleo pode sinalizar crescimento global (mais demanda por energia) → otimismo → bolsas sobem.
• Mas se a alta for por choques de oferta (ex: guerra, crise na OPEP), pode gerar temor de inflação global → bolsas caem.

# Análise Mensal

```python
#Cria uma coluna "ano-mês"
df1['ano_mes'] = df1['Date'].dt.to_period('M')
```

```python
df1.head(2)
```

```python
colunas_media = ['Close_Real_EYAC2','Close_Dolar_EYAC2','Volume_EYAC2','Price_Dolar_Petro','Volume_Petro','Índice_Ibovespa','ano_mes']

list_dates = []

#Adicionando valores do fechamento (último dia) de cada mês
for x in range (2010,2025):
    for y in range (1,13):
        list_dates.append(list(df1[(df1['Date'].dt.year==x) & (df1['Date'].dt.month == y)].iloc[-1]))

#Transformando em dataframe com colunas selecionadas
df_mean = pd.DataFrame(list_dates, columns=df1.columns)
df_mean = df_mean[colunas_media]
```

```python
df_mean = df_mean.merge(df1[['ano_mes','Value_IPCA']].drop_duplicates(),right_on='ano_mes', left_on='ano_mes',how='left')
```

```python
df_mean.tail(2)
```

```python
df_mean['Change%_Petro'] = df_mean['Price_Dolar_Petro'].pct_change() * 100
df_mean['Change%_Ibovespa'] = df_mean['Índice_Ibovespa'].pct_change() * 100
```

```python
df_mean.head(5)
```

```python
#Valor da inflação como tipo float
df_mean['Inflac_IPCA'] = df_mean['Value_IPCA'].replace({'%':''},regex=True).astype(float)
```

```python
#Adicionando coluna de inflação acumulada em 12 meses

lista_ipca = []

for i in df_mean.index:
    taxas_inflacao = [x for x in df_mean['Inflac_IPCA'].iloc[max(0, i-12):i] / 100]
    #Calculando a inflação acumulada
    inflacao_acumulada = 1
    for taxa in taxas_inflacao:
        inflacao_acumulada *= (1 + taxa)

    inflacao_acumulada -= 1
    inflacao_acumulada = inflacao_acumulada * 100  #Convertendo para porcentagem
    lista_ipca.append(inflacao_acumulada)

df_mean['IPCA_Acumulado'] = lista_ipca
```
