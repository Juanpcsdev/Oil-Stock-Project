```python
# Supondo que df1 seja o seu DataFrame original
# Transforme o DataFrame usando melt, excluindo o índice Ibovespa
df_melted = df1.melt(id_vars=['Date'], value_vars=['Price_Dolar_Petro', 'Close_Dolar_EYAC2'], 
                     var_name='Variable', value_name='Value')

# Crie os subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Primeiro gráfico: Price_Dolar_Petro
sns.lineplot(data=df_melted[df_melted['Variable'] == 'Price_Dolar_Petro'], 
             x='Date', y='Value', ax=axes[0], palette="tab10", linewidth=2.5)
axes[0].set_title('Price Dolar Petro')

# Segundo gráfico: Close_Dolar_EYAC2
sns.lineplot(data=df_melted[df_melted['Variable'] == 'Close_Dolar_EYAC2'], 
             x='Date', y='Value', ax=axes[1], palette="tab10", linewidth=2.5)
axes[1].set_title('Close Dolar EYAC2')

# Ajuste o layout
plt.tight_layout()

# Exibir os gráficos
plt.show()
```
