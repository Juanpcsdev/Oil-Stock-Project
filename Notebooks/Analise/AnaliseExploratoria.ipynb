1. Influência Direta: Ações Ligadas ao Setor de Petróleo
• O Ibovespa é composto por várias empresas, e Petrobras (PETR3, PETR4) tem um peso enorme no índice.
• Como a Petrobras é uma petroleira, o preço do barril de petróleo influencia diretamente seu lucro.
• Alta no petróleo → tende a aumentar lucro da Petrobras → ações sobem → Ibovespa sobe.
• Queda no petróleo → pode reduzir lucro → ações caem → impacto negativo no Ibovespa.
⸻
2. Relação com o Câmbio
• O petróleo é cotado em dólar, e o Brasil é um exportador de petróleo.
• Quando o petróleo sobe, pode haver:
• Entrada de dólares no Brasil (via exportações)
• Valorização do real (ou menor pressão de desvalorização)
• Isso pode impactar positivamente o Ibovespa, especialmente empresas que se beneficiam de dólar mais baixo (como varejo e consumo interno).
⸻
3. Impacto na Inflação e Juros
• O petróleo afeta o custo dos combustíveis (gasolina, diesel), que influenciam o IPCA (inflação).
• Alta no petróleo → aumento da inflação → pode levar o Banco Central a subir juros → impacto negativo no Ibovespa.
• Então, o petróleo também afeta o ciclo de política monetária indiretamente.
⸻
4. Relação Global e Sentimento de Mercado
• Alta no petróleo pode sinalizar crescimento global (mais demanda por energia) → otimismo → bolsas sobem.
• Mas se a alta for por choques de oferta (ex: guerra, crise na OPEP), pode gerar temor de inflação global → bolsas caem.

##### Correlação Barril de Petróleo e Ação em Dólar

```python
df1[['Price_Dolar_Petro', 'Close_Dolar_EYAC2']].corr(method='pearson')
```

```python
plt.figure(figsize=(10, 6))
x = df1['Price_Dolar_Petro']
y = df1['Close_Dolar_EYAC2']
plt.scatter(x, y)
plt.title('Correlação entre os preços do barril de petróleo e a ação em Dólar')
plt.xlabel('Preço Barril')
plt.ylabel('Preço Ação')

#Linha de tendência
z = np.polyfit(x,y,1)
p = np.polyval(z, x)
plt.plot(x, p, color='red')

plt.show()
```

A correlação do preço do barril do petróleo e ação estão diretamento proporcionais, quando o preço do barril do petróleo aumenta a da ação também irá aumentar
