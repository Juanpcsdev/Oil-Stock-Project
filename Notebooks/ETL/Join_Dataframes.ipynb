```python
import pandas as pd
```

```python
df1 = pd.read_csv('../../camada_prata/eyac2_clean.csv', delimiter = ',')

display(df1)
```

```python
df2 = pd.read_csv('../../camada_prata/Crude_Oil_clean.csv', delimiter = ',')

display(df2)
```

```python
df3 = pd.read_csv('../../camada_prata/ibovespa_clean.csv')

display(df3)
```

```python
df4 = pd.read_csv('../../camada_prata/ipca_clean.csv')

display(df4)
```

```python
df_merge1 = df1.merge(df2, left_on= "Date", right_on= "Date", suffixes= ("_EYAC2", "_Petro"))

display(df_merge1)
```

```python
df_merge2 = df_merge1.merge(df3[["Date","Valor","Change %","Year-Month"]], left_on= "Date", right_on= "Date", suffixes= ("","_Ibovespa"))

df_merge2.head()
```

```python
df_merge2.head()
```

```python
df_merge3 = df_merge2.merge(df4[["Year-Month", "Value", "Inflac_IPCA", "IPCA_Acumulado"]], left_on= "Year-Month", right_on= "Year-Month")

df_merge3.head(3)
```

```python
df_merge3 = df_merge3.rename(columns={"Close":"Close_Real_EYAC2","Valor_Dolar":"Close_Dolar_EYAC2",
                                     "Price":"Price_Dolar_Petro","Change %":"Change%_Petro",  
                                     "Valor":"Índice_Ibovespa", "Change %_Ibovespa":"Change%_Ibovespa", 
                                     "Value":"Value_IPCA"})

df_merge3.head(3)
```

```python
df_merge3 = df_merge3.drop(columns=['Datetime',"Change%_Ibovespa","Change%_Petro"])
```

```python
df_merge3['Date'] = pd.to_datetime(df_merge3['Date'])
```

```python
df_merge3.info()
```

```python
list_dates = []

#Adicionando valores do fechamento (último dia) de cada mês
for x in range (2010,2025):
    for y in range (1,13):
        list_dates.append(list(df_merge3[(df_merge3['Date'].dt.year==x) & (df_merge3['Date'].dt.month == y)].iloc[-1]))
```

```python
#Transformando em dataframe com colunas selecionadas
df_main = pd.DataFrame(list_dates, columns=df_merge3.columns)
#df_mean = df_mean[colunas_media]
```

```python
df_main.head()
```

```python
# Exemplo de valores percentuais anuais
variacao_ibovespa = df_main['Índice_Ibovespa'].pct_change() # Variação percentual do Ibovespa
inflacao_ipca = df_main['IPCA_Acumulado']       # Inflação IPCA
variacao_acao = df_main['Close_Real_EYAC2'].pct_change()      # Variação percentual da sua ação

# Cálculo da taxa de juros real do Ibovespa e da ação
df_main['taxa_real_ibovespa'] = ((1 + variacao_ibovespa) / (1 + inflacao_ipca / 100) - 1) * 100
df_main['taxa_real_acao'] = ((1 + variacao_acao) / (1 + inflacao_ipca / 100) - 1) * 100

```

```python
df_main.isna().sum()
```

```python
df_main = df_main.fillna(0)
```
