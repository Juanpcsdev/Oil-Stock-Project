```python
import pandas as pd
```

```python
df1 = pd.read_csv('../../camada_prata/eyac2_clean.csv', delimiter = ',')

display(df1)
```

```python
df2 = pd.read_csv('../../camada_prata/Crude_Oil_clean.csv', delimiter = ',')

display(df2)
```

```python
df3 = pd.read_csv('../../camada_prata/ibovespa_clean.csv')

display(df3)
```

```python
df4 = pd.read_csv('../../camada_prata/ipca_clean.csv')

display(df4)
```

```python
df_merge1 = df1.merge(df2, left_on= "Date", right_on= "Date", suffixes= ("_EYAC2", "_Petro"))

display(df_merge1)
```

```python
df_merge2 = df_merge1.merge(df3[["Date","Valor","Change %","Year-Month"]], left_on= "Date", right_on= "Date", suffixes= ("","_Ibovespa"))

df_merge2.head()
```

```python
df_merge3 = df_merge2.merge(df4[["Year-Month", "Value"]], left_on= "Year-Month", right_on= "Year-Month")

df_merge3.head(3)
```

```python
df_merge3 = df_merge3.rename(columns={"Close":"Close_Real_EYAC2","Valor_Dolar":"Close_Dolar_EYAC2",
                                     "Price":"Price_Dolar_Petro","Change %":"Change%_Petro",  
                                     "Valor":"Índice_Ibovespa", "Change %_Ibovespa":"Change%_Ibovespa", 
                                     "Value":"Value_IPCA"})

df_merge3.head(3)
```

```python
df_merge3 = df_merge3.drop(columns=["Year-Month"])
```

```python
df_merge3.info()
```

```python
colunas =['Close_Real_EYAC2', 'Volume_EYAC2', 'Close_Dolar_EYAC2',
       'Price_Dolar_Petro', 'Volume_Petro', 'Change%_Petro', 'Índice_Ibovespa',
       'Change%_Ibovespa', 'Value_IPCA']

df_merge3[colunas] = df_merge3[colunas].astype("str")
```

```python
df_merge3.info()
```

```python
df_merge3[colunas] = df_merge3[colunas].map(lambda x: x.replace(".",","))
```

```python
df_merge3.head(3)
```

```python
df_merge3.to_csv('../../camada_ouro/Join_dataframes_new_end.csv', index = False)
```
