```python
import pandas as pd
```

```python
df1 = pd.read_csv('../../camada_prata/eyac2_clean.csv', delimiter = ',')

display(df1)
```

```python
df2 = pd.read_csv('../../camada_prata/Crude_Oil_clean.csv', delimiter = ',')

display(df2)
```

```python
df3 = pd.read_excel('../../camada_prata/ibovespa_clean.xlsx')

display(df3)
```

```python
df_merge1 = df1.merge(df2, left_on= "Date", right_on= "Date", suffixes= ("_EYAC2", "_Petro"))

display(df_merge1)
```

```python
df_merge2 = df_merge1.merge(df3, left_on= "Date", right_on= "Date", suffixes= ("","_Ibovespa"))

df_merge2.head()
```

```python
df_merge2.shape
```

```python
df_merge2 = df_merge2.rename(columns={"Close":"Close_Real_EYAC2","Valor_Dolar":"Close_Dolar_EYAC2",
                                     "Price":"Price_Dolar_Petro","Dia":"Day_Ibovespa",
                                     "Change %":"Change%_Petro","Mês":"Month_Ibovespa",
                                     "Ano":"Year_Ibovespa", "Valor":"Value_Ibovespa", 
                                     "Change %_Ibovespa":"Change%_Ibovespa"})

df_merge2.head()
```
