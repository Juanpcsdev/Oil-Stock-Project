```python
import pandas as pd

path_file = r"..\..\camada_bronze\ipca_mensal.xlsx"
df = pd.read_excel(path_file)

df = df.applymap(lambda x: str(x).replace('"', '').replace(',', '.'))
df.columns = df.columns.str.replace('"', '')

display(df)
```

```python
df.drop(columns=["Total"], inplace=True)
df.rename(columns={
    "Ano": "Year", "Jan": "Jan", "Fev": "Feb", "Mar": "Mar", "Abr": "Apr",
    "Mai": "May", "Jun": "Jun", "Jul": "Jul", "Ago": "Aug", "Set": "Sep",
    "Out": "Oct", "Nov": "Nov", "Dez": "Dec"
}, inplace=True)

df_gold = df.melt(id_vars=["Year"], var_name="Month", value_name="Value")

month_map = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}
df_gold["Month"] = df_gold["Month"].map(month_map)

df_gold["Year-Month"] = df_gold["Year"].astype(str) + "-" + df_gold["Month"].astype(str)

df_gold = df_gold.sort_values(by=["Year", "Month"], ascending=[True, True])

df_gold = df_gold[["Year", "Month", "Year-Month", "Value"]]
```

```python
df_gold = df_gold.reset_index()
```

```python
df_gold.drop(columns=['index'])
```

```python
#Valor da inflação como tipo float
df_gold['Inflac_IPCA'] = df_gold['Value'].replace({'%':''},regex=True).astype(float)
```

```python
#Adicionando coluna de inflação acumulada em 12 meses

lista_ipca = []

for i in df_gold.index:
    taxas_inflacao = [x for x in df_gold['Inflac_IPCA'].iloc[max(0, i-12):i] / 100]
    #Calculando a inflação acumulada
    inflacao_acumulada = 1
    for taxa in taxas_inflacao:
        inflacao_acumulada *= (1 + taxa)

    inflacao_acumulada -= 1
    inflacao_acumulada = inflacao_acumulada * 100  #Convertendo para porcentagem
    lista_ipca.append(inflacao_acumulada)

df_gold['IPCA_Acumulado'] = lista_ipca
```

```python
output_path_gold = r"..\..\camada_prata\ipca_clean.csv"
df_gold.to_csv(output_path_gold, sep=",", index=False)

print(f"Arquivo salvo em: {output_path_gold}")
display(df_gold)
```
