```python
import pandas as pd

path_file = r"..\..\camada_bronze\ipca_mensal.xlsx"
df = pd.read_excel(path_file)

df = df.applymap(lambda x: str(x).replace('"', '').replace(',', '.'))
df.columns = df.columns.str.replace('"', '')

output_path = r"..\..\camada_prata\ipca_mensal_silver.xlsx"
df.to_excel(output_path, index=False)

print(f"Arquivo salvo em: {output_path}")
display(df)
```

```python
path_file_silver = r"..\..\camada_prata\ipca_mensal_silver.xlsx"
df_silver = pd.read_excel(path_file_silver)

df_silver.drop(columns=["Total"], inplace=True)
df_silver.rename(columns={
    "Ano": "Year", "Jan": "Jan", "Fev": "Feb", "Mar": "Mar", "Abr": "Apr",
    "Mai": "May", "Jun": "Jun", "Jul": "Jul", "Ago": "Aug", "Set": "Sep",
    "Out": "Oct", "Nov": "Nov", "Dez": "Dec"
}, inplace=True)

df_gold = df_silver.melt(id_vars=["Year"], var_name="Month", value_name="Value")

month_map = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}
df_gold["Month"] = df_gold["Month"].map(month_map)

df_gold["Year-Month"] = df_gold["Year"].astype(str) + "-" + df_gold["Month"].astype(str)

df_gold = df_gold.sort_values(by=["Year", "Month"], ascending=[False, True])

df_gold = df_gold[["Year", "Month", "Year-Month", "Value"]]

output_path_gold = r"..\..\camada_ouro\ipca_final_gold.csv"
df_gold.to_csv(output_path_gold, sep=",", index=False)

print(f"Arquivo salvo em: {output_path_gold}")
display(df_gold)
```
