```python
#Importação de bibliotecas
import pandas as pd
```

```python
#Leitura do arquivo CSV
df = pd.read_csv("camada bronze/acao_petroleo.csv", delimiter=";")
```

```python
#Exibição do Dataframe
display(df)
```

```python
#Verifica se há valores nulos
df.isna().sum()
```

```python
#Preenche valores ausentes para 0
df = df.fillna(0)
```

```python
#Remove linhas duplicadas
df = df.drop_duplicates()
```

```python
#Limpa espaços em branco nas colunas com string
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#Identificando os tipos de cada coluna
df.info()
```

```python
#Altera a vírgula para ponto
colunas = ["Open","High","Low","Close","Volume"]

df[colunas] = df[colunas].map(lambda x: x.replace(",","."))
```

```python
#Converte as colunas do tipo objeto para numérico(float)
df[colunas] = df[colunas].apply(lambda x: pd.to_numeric(x))
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
#Verificando novamente os tipos das colunas modificadas
df.info()
```

```python
#Renomeia a coluna Date para Datetime, condizendo melhor com o valor da coluna
df = df.rename(columns={"Date":"Datetime"})
```

```python
#Cria uma nova coluna com o valor apenas da data da coluna Datetime
df["Date"] = df["Datetime"].dt.date
```

```python
#Reordena as colunas para que a coluna Date seja a primeira
last_column = df.columns[-1]
df = df[[last_column] + [col for col in df.columns if col != last_column]]
```

```python
#Exibição do Dataframe
display(df)
```
